```python
# Since this is only a limit amount of data, we opted to remove these
df_updated = df.dropna(axis=0)
```

```python
# Check if all non-values have been properly removed
df_updated.info()
```

```python
# Now we can turn categorical values into numerical values (sex & region)
# Male = 0, Female = 1
pd.options.mode.chained_assignment = None  # default='warn'
df_updated['SEX'] = df_updated['SEX'].map( {'F': 1, 'M': 0} ).astype(int)
```

```python
# Same for region
df_updated['REGION'] = df_updated['REGION'].map({"Flanders": 0, "Wallonia": 1, "Brussels": 2}).astype(int)
```
